```python

# ZOMATO EXPLORATORY DATA ANALYSIS PROJECT
# Project By: Khushi Upadhayay


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from IPython.display import Image, display

# Banner Image
display(Image(url='https://upload.wikimedia.org/wikipedia/commons/7/75/Zomato_logo.png', width=350))

print("="*70)
print("🍽️ ZOMATO EXPLORATORY DATA ANALYSIS PROJECT")
print("👩‍💻 Project By: Khushi Upadhayay")
print("="*70)
```

```python
# LOAD DATA

df = pd.read_csv("Zomato.csv", header=1)

print("\nDataset Shape:", df.shape)
print("\nColumns:", df.columns)
print("\nMissing Values:\n", df.isnull().sum())
print("\nDuplicate Rows:", df.duplicated().sum())
```

```python
# CLEANING

df.drop_duplicates(inplace=True)

df['Rate'] = df['Rate'].astype(str)
df['Rate'] = df['Rate'].str.replace('/5', '', regex=False)
df['Rate'] = pd.to_numeric(df['Rate'], errors='coerce')

print("\nData Cleaning Completed")
```

```python
# CHART 1 — ONLINE ORDER

plt.figure(figsize=(6,4))
sns.countplot(x='Online_order', data=df)
plt.title("Online Order Distribution")
plt.show()
```

```python
# CHART 2 — TABLE BOOKING

plt.figure(figsize=(6,4))
sns.countplot(x='Book_table', data=df)
plt.title("Table Booking Distribution")
plt.show()
```

```python
# CHART 3 — PIE CHART

df['Listed_in(type)'].value_counts().plot.pie(
    autopct='%1.1f%%',
    figsize=(7,7)
)
plt.title("Restaurant Type Distribution")
plt.ylabel("")
plt.show()
```

```python
# CHART 4 — RATING DISTRIBUTION

plt.figure(figsize=(8,4))
sns.histplot(df['Rate'], bins=20, kde=True)
plt.title("Rating Distribution")
plt.show()
```

```python
# CHART 5 — TOP 10 RESTAURANTS BY VOTES

top_votes = df.nlargest(10, 'Votes')

plt.figure(figsize=(10,6))
sns.barplot(data=top_votes, x='Votes', y='Name')
plt.title("Top 10 Restaurants by Votes")
plt.show()
```

```python
# CHART 6 — COST DISTRIBUTION

plt.figure(figsize=(8,4))
sns.histplot(df['Approx_cost(for two people)'], bins=25, kde=True)
plt.title("Cost Distribution")
plt.show()
```

```python
# CHART 7 — COST VS RATING

plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df,
    x='Approx_cost(for two people)',
    y='Rate',
    hue='Online_order'
)
plt.title("Cost vs Rating")
plt.show()
```

```python
# CHART 8 — BOXPLOT

plt.figure(figsize=(7,4))
sns.boxplot(x='Online_order', y='Rate', data=df)
plt.title("Online Order vs Rating")
plt.show()
```

```python
# CHART 9 — VIOLIN PLOT

plt.figure(figsize=(7,4))
sns.violinplot(x='Book_table', y='Rate', data=df)
plt.title("Table Booking vs Rating")
plt.show()
```

```python
# CHART 10 — HEATMAP

numeric = df[['Votes', 'Rate', 'Approx_cost(for two people)']]

plt.figure(figsize=(6,4))
sns.heatmap(numeric.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# CHART 11 — KDE PLOT

plt.figure(figsize=(8,4))
sns.kdeplot(df['Rate'].dropna(), fill=True)
plt.title("Density of Ratings")
plt.show()
```

```python
# CHART 12 — PAIRPLOT (ADVANCED)

sns.pairplot(
    df[['Votes','Rate','Approx_cost(for two people)']].dropna()
)
plt.show()
```
